```python
from sklearn.datasets import load_diabetes
```

```python
d = load_diabetes()
d
```

```python
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d[:2]
```

```python
X = pd.DataFrame(d.data)
X.rename(columns ={0:'age',1:'sex',2:'bmi',3:'bp',4:'s1',5:'s2',6:'s3',7:'s4',8:'s4',9:'s5'},inplace=True)
X
```

```python
t = d.target
y = pd.DataFrame(t,columns=['Target'])
y
```

```python
plt.figure(figsize=(250,500))
sns.heatmap(data=d.data)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
X_t = X.head()
X_t
```

```python
y_pred = lr.predict(X_t)
y_pred
```

```python
y.head()
```

```python
lr.score(X,y)
```
